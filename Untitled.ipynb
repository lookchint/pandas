```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

```python
path = 'AT-33-F_SX(TCombo)_6.125_ULog_Envi.las'
with open(path, encoding="ISO-8859-1") as f:
    header_section = False
    data_section = False
    headers = []
    data = []
    for line in f:
        if header_section and ('~' in line):
            header_section = False

        if '~curve information' in line.lower():
            header_section = True

        if '~A' in line:
            data_section = True

        if header_section and ('#' != line[0]) and ('~' != line[0]):
            headers.append(line.split()[0].split('.')[0].strip())
    
        if data_section and ('#' != line[0]) and ('~' != line[0]):
            data.append(re.split('\s+', line.strip()))

df_file_las = pd.DataFrame(data, columns=headers)
df_file_las = df_file_las.apply(pd.to_numeric)
```
